```python
#Select features 

# write list of features we want to use
features = ["age", "education_level", "marital-status", "occupation", "race", "sex", "workclass", "hours-per-week"]

# Set X
X_temp = data[features]

# Convert text to numeric
X = pd.get_dummies(X_temp)

# Set y
y_temp = data["income"]

# Convert text to numeric
y = pd.get_dummies(y_temp)
y = y.drop('<=50K', axis=1) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 42)
```

```python
import xgboost as xgb
```

```python
from xgboost import XGBClassifier
```

```python
# initialize model
log_reg = XGBClassifier()

# fit model
log_reg.fit(X_train, y_train)

# metrics (accuracy)
acc = log_reg.score(X_test, y_test)


# predictions
y_pred = log_reg.predict(X_test)

acc
```

```python
metrics.precision_score(y_test, y_pred)
```

```python
metrics.recall_score(y_test, y_pred)
```
